### 1. Autodatan yhdistäminen

Tee funktio autot(url1, url2), joka tekee järjestyksessä seuraavaa:
* lataa datat DataFrameihin url1:sta ja url2:sta, joissa erotinmerkkinä on kaksoispiste (';'),
* yhdistää uuteen DataFrameen autoihin liittyvät datat url1:stä ja url2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla),
* tekee juuri yhdistetystä DataFramesta Serieksen, josta käy ilmi kuinka monta automerkkiä datassa on (automerkki on sarakkeessa 'Cn'),
* tulostaa em. Serieksestä viisi eniten esiintyvää automerkkiä.

Esimerkiksi, autot(url1, url2):

|OCTAVIA   | 396|
|FABIA     | 377|
|RAPID     | 197|
|SUPERB    | 159|
|KAROQ     | 148|
|Name: Cn, dtype: int64 | |

```python
# Testaa funktiotasi:
url1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
url2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'

def autot(url1, url2):
    # Kirjoita vastauksesi tähän
    import pandas as pd


    # lataa datat DataFrameihin url1:sta ja url2:sta, joissa erotinmerkkinä on kaksoispiste (';')

    df1 = pd.read_csv(url1,
                 sep= ';')

    df2 = pd.read_csv(url2,
                 sep= ';')


    # yhdistää uuteen DataFrameen autoihin liittyvät datat url1:stä ja url2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla)
    mergedDf = pd.merge(df1, df2, how='outer')


    # tekee juuri yhdistetystä DataFramesta Serieksen, josta käy ilmi kuinka monta automerkkiä datassa on (automerkki on sarakkeessa 'Cn')
    result = mergedDf['Cn'].value_counts()


    # tulostaa em. Serieksestä viisi eniten esiintyvää automerkkiä.
    tulos = result.head()
    
    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)
    


autot(url1, url2)
```

### 2. Autodatan yhdistäminen ja ryhmittely

Tee funktio autot(url_a1, url_a2, url_b1, url_b2), joka tekee järjestyksessä seuraavaa:
* lataa datat DataFrameihin osoitteista (url_a1, url_a2, url_b1, url_b2), joissa erotinmerkkinä on kaksoispiste (';'),
* yhdistää uuteen DataFrameen 'dfa' autoihin liittyvät datat url_a1:sta ja url_a2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla),
* yhdistää uuteen DataFrameen 'dfb' autoihin liittyvät datat url_b1:sta ja url_b2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla),
* yhdistää DataFramet 'dfa' ja 'dfb' yhdeksi DataFrameksi 'df' sarakkeen 'ID' perusteella,
* jos DataFramessa 'df' on sarakkeessa auton moottorin teho 'ep (KW)' nan arvoja, korvataan ne lukuarvolla nolla,
* tuottaa DataFramen 'tulos', johon on laskettu autojen keskimääräiset tehot sarakkeesta 'ep (KW)' ryhmiteltynä auton valmistajan mukaan 'Mh' (tutki tarvittaessa groupby [dokumentaatiosta](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.groupby.html?highlight=groupby#pandas.DataFrame.groupby) kuinka groupby palauttaa indeksin),
* järjestää DataFramen 'tulos' rivit keskimääräisen tehon mukaan 'ep (KW)' suurimmasta pienimpään.
* tulostaa edellä jäjestetystä DataFramesta viisi ensimmäistä riviä.

Tässä tehtävässä:
* url_a1=https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv
* url_a2=https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv
* url_b1=https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv
* url_b2=https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv

Esimerkiksi, autot(url_a1, url_a2, url_b1, url_b2):

|:---|:---|:---|
|6 |           TESLA | 386.000000|
|0 |          AA-IVA | 267.750000|
|2 | FORD WERKE GMBH |  89.741007|
|5 |           SKODA |  81.372515|
|3 |             KIA |  77.000000|

```python
def autot(url_a1, url_a2, url_b1, url_b2):
    # Kirjoita vastauksesi tähän
    # lataa datat DataFrameihin osoitteista (url_a1, url_a2, url_b1, url_b2), joissa erotinmerkkinä on kaksoispiste (';')
    df1 = pd.read_csv(url_a1,
                 sep= ';')

    df2 = pd.read_csv(url_a2,
                 sep= ';')

    df3 = pd.read_csv(url_b1,
                 sep= ';')

    df4 = pd.read_csv(url_b2,
                 sep= ';')

    # yhdistää uuteen DataFrameen 'dfa' autoihin liittyvät datat url_a1:sta ja url_a2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla),
    dfa = pd.concat([df1, df2])
    # yhdistää uuteen DataFrameen 'dfb' autoihin liittyvät datat url_b1:sta ja url_b2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla),
    dfb = pd.concat([df3, df4])


    # yhdistää DataFramet 'dfa' ja 'dfb' yhdeksi DataFrameksi 'df' sarakkeen 'ID' perusteella,
    df = pd.merge(dfa, dfb, how='inner', left_on='ID', right_on='ID')


    # jos DataFramessa 'df' on sarakkeessa auton moottorin teho 'ep (KW)' nan arvoja, korvataan ne lukuarvolla nolla
    df['ep (KW)'] = df['ep (KW)'].fillna(int(0))


    # tuottaa DataFramen 'tulos', johon on laskettu autojen keskimääräiset tehot sarakkeesta 'ep (KW)' ryhmiteltynä auton valmistajan mukaan 'Mh'
    tulos1 = df.groupby('Mh', as_index=False)['ep (KW)'].mean()
    
    # järjestää DataFramen 'tulos' rivit keskimääräisen tehon mukaan 'ep (KW)' suurimmasta pienimpään ja tulostaa edellä jäjestetystä DataFramesta viisi ensimmäistä riviä.
    tulos = tulos1.sort_values(by='ep (KW)', ascending=False).head()
    
    
    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)
    
url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

autot(url_a1, url_a2, url_b1, url_b2)
```

## 3. Autodatan yhdistäminen ja ristiintaulukointi

Tee funktio autot(url_a1, url_a2, url_b1, url_b2), joka tekee järjestyksessä seuraavaa:
* lataa datat DataFrameihin osoitteista (url_a1, url_a2, url_b1, url_b2), joissa erotinmerkkinä on kaksoispiste (';'),
* yhdistää uuteen DataFrameen 'dfa' autoihin liittyvät datat url_a1:sta ja url_a2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla),
* yhdistää uuteen DataFrameen 'dfb' autoihin liittyvät datat url_b1:sta ja url_b2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla),
* yhdistää DataFramet 'dfa' ja 'dfb' yhdeksi DataFrameksi 'df' sarakkeen 'ID' perusteella,
* jos sarakkeessa 'Ft' on merkkijono 'DIESEL', korvataan se merkkijonolla 'Diesel',
* jos sarakkeessa 'Ft' on merkkijono 'PETROL', korvataan se merkkijonolla 'Petrol',
* tekee uuden DataFramen hyödyntäen [crosstab funktiota](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.crosstab.html) ja sopivaa numpy [funktiota](https://numpy.org/doc/stable/reference/), jossa on laskettu esimerkkituloksen mukaisesti keskimääräiset päästöt sarakkeesta 'Enedc (g/km)' autonvalmistaja kohtaisesti (sarake 'Mh') polttoainetyypeittäin (sarake 'Ft'),
* esittää edellä kuvatusta DataFramesta keskimääräiset päästöt yhden desimaalin tarkkuudella,
* valitsee uuteen DataFrameen sarakkeet 'Dielsel' ja 'Petrol',
* tulostaa edellä tuotetusta DataFramesta viisi ensimmäistä riviä.

Tässä tehtävässä:
* url_a1=https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv
* url_a2=https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv

```python
def autot(url_a1, url_a2, url_b1, url_b2):
    # Kirjoita vastauksesi tähän
    import pandas as pd
    import numpy as np

    # lataa datat DataFrameihin osoitteista (url_a1, url_a2, url_b1, url_b2), joissa erotinmerkkinä on kaksoispiste (';')

    df1 = pd.read_csv(url_a1,
                 sep= ';')

    df2 = pd.read_csv(url_a2,
                 sep= ';')

    df3 = pd.read_csv(url_b1,
                 sep= ';')

    df4 = pd.read_csv(url_b2,
                 sep= ';')


    # asetetaan datatyypit samoiksi dfb-dataframeihin
    df3 = df3.astype('object')
    df4 = df4.astype('object')


    # yhdistää uuteen DataFrameen 'dfa' autoihin liittyvät datat url_a1:sta ja url_a2:sta 
    # (tutki tarvittaessa mitä dataa urleista on saatavilla)
    dfa = pd.merge(df1, df2, how='outer')
    dfb = pd.merge(df3, df4, how='outer')


    # jos sarakkeessa 'Ft' on merkkijono 'DIESEL', korvataan se merkkijonolla 'Diesel',
    dfb['Ft'] = dfb['Ft'].str.replace('DIESEL', 'Diesel')
    # jos sarakkeessa 'Ft' on merkkijono 'PETROL', korvataan se merkkijonolla 'Petrol',
    dfb['Ft'] = dfb['Ft'].str.replace('PETROL', 'Petrol')


    # tekee uuden DataFramen hyödyntäen crosstab funktiota ja sopivaa numpy funktiota,
    # jossa on laskettu esimerkkituloksen mukaisesti keskimääräiset päästöt sarakkeesta 'Enedc (g/km)' autonvalmistaja kohtaisesti (sarake 'Mh') polttoainetyypeittäin (sarake 'Ft')
    # esittää keskimääräiset päästöt yhden desimaalin tarkkuudella,
    crossDf = pd.crosstab(index= dfa['Mh'], columns= dfb['Ft'], values= dfb['Enedc (g/km)'], aggfunc= 'mean').round(1)


    # valitsee uuteen DataFrameen sarakkeet 'Dielsel' ja 'Petrol'
    resultDf = crossDf.drop(['NG', 'Electric'], axis=1)

    # tulostaa edellä tuotetusta DataFramesta viisi ensimmäistä riviä.
    tulos = resultDf.head().astype(str)
    
    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)
    
# Testaa funktiotasi:
url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

autot(url_a1, url_a2, url_b1, url_b2)
```

### 4. Autodatan yhdistäminen ja ryhmittely monitasoiseksi DataFrameksi

```python
def autot(url_a1, url_a2, url_b1, url_b2):
    # Kirjoita vastauksesi tähän
    import pandas as pd
    import numpy as np

    # lataa datat DataFrameihin osoitteista (url_a1, url_a2, url_b1, url_b2), joissa erotinmerkkinä on kaksoispiste (';')

    df1 = pd.read_csv(url_a1,
                 sep= ';')

    df2 = pd.read_csv(url_a2,
                 sep= ';')

    df3 = pd.read_csv(url_b1,
                 sep= ';')

    df4 = pd.read_csv(url_b2,
                 sep= ';')

    pd.options.mode.use_inf_as_na = True


    # asetetaan datatyypit samoiksi dfb-dataframeihin
    df3 = df3.astype('object')
    df4 = df4.astype('object')


    # yhdistää uuteen DataFrameen 'dfa' autoihin liittyvät datat url_a1:sta ja url_a2:sta 
    # (tutki tarvittaessa mitä dataa urleista on saatavilla)
    dfa = pd.merge(df1, df2, how='outer')
    dfb = pd.merge(df3, df4, how='outer')


    # jos sarakkeessa 'Ft' on merkkijono 'DIESEL', korvataan se merkkijonolla 'Diesel',
    dfb['Ft'] = dfb['Ft'].str.replace('DIESEL', 'Diesel')
    # jos sarakkeessa 'Ft' on merkkijono 'PETROL', korvataan se merkkijonolla 'Petrol',
    dfb['Ft'] = dfb['Ft'].str.replace('PETROL', 'Petrol')


    # tekee uuden DataFramen, jossa on laskettu moottorien keskiarvotilavuudet (sarake 'ec (cm3)') ryhmiteltynä sarakkeiden 'Mh' ja 'Ft' mukaan,
    crossDf = pd.crosstab(index= dfa['Mh'], columns= dfb['Ft'], values= dfb['ec (cm3)'], aggfunc= 'mean', dropna=False).round(0)


    # esittää moottorien keskiarvotilavuudet kokonaislukujen tarkkuudella
    crossDf2 = crossDf.astype('Int32')

    # tiputetaan 'NG'-sarake pois
    crossDf2.drop(['NG', 'Electric'], axis=1, inplace=True)

    # muodostaa monitasoisen DataFramen, jossa sarakeindekseinä ovat monikot ('ec (cm3)', 'Diesel') ja ('ec (cm3)', 'Petrol')
    # määritetään sarakkeiden nimet tupleina
    tuple_col = [('ec (cm3)', 'Diesel'),
         ('ec (cm3)', 'Petrol')]

    # vaihdetaan sarakkeiden nimet
    crossDf2.columns = tuple_col

    # muodostetaan multiindeksi-sarakkeet tupleista
    crossDf2.columns = pd.MultiIndex.from_tuples(crossDf2.columns, names=['', 'Ft'])


    # jos DataFramessa esiintyy NaN arvoja, niin DataFramen soluun kirjoitetaan merkkijono "no value"
    resultDf = crossDf2.astype(object).fillna('no value')

    tulos = resultDf.head(10)
    
    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)
    
# Testaa funktiotasi:
url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

autot(url_a1, url_a2, url_b1, url_b2)
```

```python
def autot(url_a1, url_a2, url_b1, url_b2):
    # Kirjoita vastauksesi tähän
    import pandas as pd
    import numpy as np

    # lataa datat DataFrameihin osoitteista (url_a1, url_a2, url_b1, url_b2), joissa erotinmerkkinä on kaksoispiste (';')

    df1 = pd.read_csv(url_a1,
                 sep= ';')

    df2 = pd.read_csv(url_a2,
                 sep= ';')

    df3 = pd.read_csv(url_b1,
                 sep= ';')

    df4 = pd.read_csv(url_b2,
                 sep= ';')


    # yhdistää uuteen DataFrameen 'dfa' autoihin liittyvät datat url_a1:sta ja url_a2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla),
    dfa = pd.concat([df1, df2])
    # yhdistää uuteen DataFrameen 'dfb' autoihin liittyvät datat url_b1:sta ja url_b2:sta (tutki tarvittaessa mitä dataa urleista on saatavilla),
    dfb = pd.concat([df3, df4])


    # yhdistää DataFramet 'dfa' ja 'dfb' yhdeksi DataFrameksi 'df' sarakkeen 'ID' perusteella,
    df = pd.merge(dfa, dfb, how='inner', left_on='ID', right_on='ID')


    # jos sarakkeessa 'Ft' on merkkijono 'DIESEL', korvataan se merkkijonolla 'Diesel',
    df['Ft'] = df['Ft'].str.replace('DIESEL', 'Diesel')
    # jos sarakkeessa 'Ft' on merkkijono 'PETROL', korvataan se merkkijonolla 'Petrol',
    df['Ft'] = df['Ft'].str.replace('PETROL', 'Petrol')


    # tallentaa uuteen DataFrameen sarakkeet 'Enedc (g/km)', 'MS' ja 'Ft'
    df5 = df[['Enedc (g/km)', 'MS', 'Ft']]


    # korvaa sarakkeen nimen 'Enedc (g/km)' nimellä 'Avg emissions (g/km)'
    # korvaa sarakkeen nimen 'MS' nimellä 'Member State'
    # korvaa sarakkeen nimen 'Ft' nimellä 'Fuel type'

    df6 = df5.copy()
    df6.rename(columns = {
                        'Enedc (g/km)' : 'Avg emissions (g/km)',
                        'MS': 'Member State',
                        'Ft': 'Fuel type'},
                        inplace=True
                      )


    # laskee uuteen DataFrameen keskimääräiset päästöt (sarake 'Avg emissions (g/km)') maittain (sarake 'Member State') ja polttoainetyypeittäin (sarake 'Fuel type')
    table = pd.pivot_table(df6, values=['Avg emissions (g/km)'], index=['Member State'], columns=['Fuel type'])


    # jättää edellä tuotettuun monitasoiseen DataFrameen vain 'Diesel' ja 'Petrol' sarakkeet
    table.drop(columns=['Electric', 'NG'], level=1, inplace=True)


    # laskee uuteen sarakkeeseen 'D-P' kuinka paljon sarkkeen 'Diesel' arvo on suurempi kuin sarakkeen 'Petrol'
    table[('Avg emissions (g/km)', 'D-P')] = table.loc[:, ('Avg emissions (g/km)', 'Diesel')] - table.loc[:, ('Avg emissions (g/km)', 'Petrol')]


    # järjestää DataFramen sarakkeen 'D-P' mukaan laskevaan järjestykseen
    table.sort_values(by=[('Avg emissions (g/km)', 'D-P')], ascending=False, inplace=True)


    # pyöristää ja esittää DataFramen lukuarvot kokonaislukujen tarkkuudella
    table2 = table.applymap("{:.0f}".format)


    # tulostaa vastaukseen DataFramesta viisi ensimmäistä riviä.
    tulos = table2.head()
    
    # Älä kirjoita muita print komentoja ennen tulos muuttujan tulostusta.
    print(tulos)
    
# Testaa funktiotasi:
url_a1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA1.csv'
url_a2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotA2.csv'
url_b1='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB1.csv'
url_b2='https://student.labranet.jamk.fi/~huoptu/datananal_s20/kerta4/harjoitukset/autotB2.csv'

autot(url_a1, url_a2, url_b1, url_b2)
```
