```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd

# crawler settings.py
download_delay = 0.5
concurrent_requests_per_domain = 2
headers = {'Accept-Encoding': 'identity'}

pitch_titles = []
all_html_data = []
URL_list = []


# create URLs to list
for x in range(1100):
    url = 'https://invest.invesdor.com/en/pitches/'
    x+=1
    URL_list.append(str(url) + str(x))
    

# read all data from URLs
for url in URL_list:
    sleep(download_delay)
    resp = requests.get(url, headers=headers)
    soup = BeautifulSoup(resp.content, "html.parser")
    all_html_data.append(soup)


# combine all data to new soup
soup = ''.join(str(all_html_data))
all_data = BeautifulSoup(soup)


# add all titles to list
pitchTitles = []
titles = all_data.find_all('h1', attrs={'class':'pitch-title'})
for y in all_data:
    pitchTitles.append(y.string)
```
